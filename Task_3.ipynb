```python
import pandas as pd

file_path = '/content/drive/MyDrive/dataset/creditcard.csv'
try:
    df = pd.read_csv(file_path)
    print(f'Successfully loaded data from {file_path}.')
    print('\nFirst 5 rows of the DataFrame:')
    display(df.head())
except FileNotFoundError:
    print(f'Error: The file {file_path} was not found. Please ensure the path is correct and your Drive is mounted.')
except Exception as e:
    print(f'An error occurred while loading the file: {e}')
```

```python
if 'df' in locals():
    initial_rows = df.shape[0]
    df.drop_duplicates(inplace=True)
    rows_after_duplicates = df.shape[0]
    print(f'Original number of rows: {initial_rows}')
    print(f'Number of rows after removing duplicates: {rows_after_duplicates}')
    print(f'Number of duplicate rows removed: {initial_rows - rows_after_duplicates}')

    print('\nFirst 5 rows of the DataFrame after removing duplicates:')
    display(df.head())

    print('\nDataFrame Info:')
    df.info()
else:
    print('DataFrame was not loaded. Cannot perform duplicate removal.')
```

```python
if 'df' in locals():
    # Calculate value counts for the 'Class' column
    class_counts = df['Class'].value_counts()

    # Calculate percentage distribution of the classes
    class_percentages = df['Class'].value_counts(normalize=True) * 100

    print('Distribution of the \'Class\' column (Fraudulent vs. Non-Fraudulent):')
    print('\nRaw Counts:')
    print(class_counts)

    print('\nPercentage Distribution:')
    print(class_percentages)
else:
    print("DataFrame 'df' not found. Please ensure the data is loaded correctly.")
```

```python
if 'df' in locals():
    # 1. Separate features (X) and target (y)
    X = df.drop('Class', axis=1)
    y = df['Class']

    print('Original Class Distribution:')
    print(y.value_counts())

    # 2. Import SMOTE
    from imblearn.over_sampling import SMOTE

    # 3. Instantiate SMOTE
    smote = SMOTE(random_state=42)

    # 4. Apply SMOTE to the data
    X_resampled, y_resampled = smote.fit_resample(X, y)

    # 5. Print the value counts of y_resampled
    print('\nClass Distribution after SMOTE:')
    print(y_resampled.value_counts())
else:
    print("DataFrame 'df' not found. Please ensure the data is loaded correctly.")
```

```python
if 'X_resampled' in locals() and 'y_resampled' in locals():
    # 1. Import the train_test_split function
    from sklearn.model_selection import train_test_split

    # 2. Split the resampled data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(
        X_resampled, y_resampled, test_size=0.3, random_state=42, stratify=y_resampled
    )

    # 3. Print the shapes of the resulting datasets
    print(f'Shape of X_train: {X_train.shape}')
    print(f'Shape of X_test: {X_test.shape}')
    print(f'Shape of y_train: {y_train.shape}')
    print(f'Shape of y_test: {y_test.shape}')
else:
    print("Resampled data (X_resampled, y_resampled) not found. Please ensure SMOTE was applied correctly.")
```

```python
from sklearn.ensemble import RandomForestClassifier

# Instantiate RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42)

# Fit the model to the balanced training data
rf_model.fit(X_train, y_train)

print("RandomForestClassifier model trained successfully.")
```

```python
if 'rf_model' in locals() and 'X_test' in locals() and 'y_test' in locals():
    # 1. Use the trained Random Forest Classifier model to make predictions on the test features
    y_pred = rf_model.predict(X_test)

    # 2. Get the probability estimates for the positive class (class 1)
    y_pred_proba = rf_model.predict_proba(X_test)[:, 1]

    # 3. Import the classification_report and roc_auc_score functions
    from sklearn.metrics import classification_report, roc_auc_score

    # 4. Generate and print the classification report
    print('Classification Report:')
    print(classification_report(y_test, y_pred))

    # 5. Calculate and print the ROC AUC score
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    print(f'ROC AUC Score: {roc_auc:.4f}')
else:
    print("Model or test data not found. Please ensure the model is trained and data is split correctly.")
```

```python
if 'y_test' in locals() and 'y_pred' in locals():
    # 1. Import the confusion_matrix function from sklearn.metrics and seaborn as sns, and matplotlib.pyplot as plt
    from sklearn.metrics import confusion_matrix
    import seaborn as sns
    import matplotlib.pyplot as plt

    # 2. Compute the confusion matrix using y_test and y_pred
    cm = confusion_matrix(y_test, y_pred)

    # 3. Create a heatmap of the confusion matrix using sns.heatmap()
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['Non-Fraudulent', 'Fraudulent'],
                yticklabels=['Non-Fraudulent', 'Fraudulent'])

    # 4. Label the x-axis as 'Predicted' and the y-axis as 'Actual'
    plt.xlabel('Predicted')
    plt.ylabel('Actual')

    # 5. Add a title to the plot, such as 'Confusion Matrix'
    plt.title('Confusion Matrix')

    # 6. Display the plot using plt.show()
    plt.show()
else:
    print("Test actuals (y_test) or predicted values (y_pred) not found. Please ensure the model evaluation was performed correctly.")
```
